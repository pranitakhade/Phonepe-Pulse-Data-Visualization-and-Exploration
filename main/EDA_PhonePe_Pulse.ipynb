```python
directory_path = "/content/pulse/data/aggregated/transaction/country/india/state"

result = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Transaction_type': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

states = os.listdir(directory_path)

for state in states:
    state_path = os.path.join(directory_path, state)
    years = os.listdir(state_path)

    for year in years:
        year_path = os.path.join(state_path, year)
        json_files = os.listdir(year_path)

        for json_file in json_files:
            json_path = os.path.join(year_path, json_file)

            with open(json_path, 'r') as json_data:
                data = json.load(json_data)

                for transaction in data['data']['transactionData']:
                    name = transaction['name']
                    count = transaction['paymentInstruments'][0]['count']
                    amount = transaction['paymentInstruments'][0]['amount']

                    result['Transaction_type'].append(name)
                    result['Transaction_count'].append(count)
                    result['Transaction_amount'].append(amount)
                    result['State'].append(state)
                    result['Year'].append(year)
                    result['Quarter'].append('Q' + os.path.splitext(json_file)[0])

df = pd.DataFrame(result)
```

```python
df.head()
```

```python
df.to_csv('PhonePe_States.csv', index=False)
```

```python
user_path = "/content/pulse/data/aggregated/user/country/india/state"

result = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'User_Brand': [],
    'User_Count': [],
    'User_Percentage': []
}

states_user = os.listdir(user_path)

for state in states_user:
    state_path = os.path.join(user_path, state)
    years = os.listdir(state_path)

    for year in years:
        year_path = os.path.join(state_path, year)
        json_files = os.listdir(year_path)

        for json_file in json_files:
            json_path = os.path.join(year_path, json_file)

            with open(json_path, 'r') as json_data:
                data = json.load(json_data)
                try:
                    for user in data['data']['usersByDevice']:
                        brand = user['brand']
                        count = user['count']
                        percentage = user['percentage']

                        result['User_Brand'].append(brand)
                        result['User_Count'].append(count)
                        result['User_Percentage'].append(percentage)
                        result['State'].append(state)
                        result['Year'].append(year)
                        result['Quarter'].append('Q' + os.path.splitext(json_file)[0])
                except TypeError:
                    pass

user_df = pd.DataFrame(result)
```

```python
user_df.head()
```

```python
user_df.to_csv('PhonePe_User.csv', index=False)
```

```python
map_path = "/content/pulse/data/map/transaction/hover/country/india/state"

map_result = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Map_District': [],
    'Map_Count': [],
    'Map_Amount': []
}

state_map = os.listdir(map_path)

for state in state_map:
    state_path_map = os.path.join(map_path, state)
    years_map = os.listdir(state_path_map)

    for year in years_map:
        year_path_map = os.path.join(state_path_map, year)
        json_files_map = os.listdir(year_path_map)

        for json_file_map in json_files_map:
            json_path_map = os.path.join(year_path_map, json_file_map)

            with open(json_path_map, 'r') as jsonFile_map:
                data_map = json.load(jsonFile_map)
                try:
                    for map_data in data_map['data']['hoverDataList']:
                        district = map_data['name']
                        count = map_data['metric'][0]['count']
                        amount = map_data['metric'][0]['amount']

                        map_result['Map_District'].append(district)
                        map_result['Map_Count'].append(count)
                        map_result['Map_Amount'].append(amount)
                        map_result['State'].append(state)
                        map_result['Year'].append(year)
                        map_result['Quarter'].append('Q' + os.path.splitext(json_file_map)[0])
                except TypeError:
                    pass

map_df = pd.DataFrame(map_result)
```

```python
map_df.head()
```

```python
map_df.to_csv("PhonePe_Map_Transaction.csv",index = False)
```

```python
map_user_path = "/content/pulse/data/map/user/hover/country/india/state"

map_user_result = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'District': [],
    'Registered_users': [],
    'App_Opens': []
}

states_map_user = os.listdir(map_user_path)

for state in states_map_user:
    state_path_map_user = os.path.join(map_user_path, state)
    years_map_user = os.listdir(state_path_map_user)

    for year in years_map_user:
        year_path_map_user = os.path.join(state_path_map_user, year)
        json_files_map_user = os.listdir(year_path_map_user)

        for json_file_map_user in json_files_map_user:
            json_path_map_user = os.path.join(year_path_map_user, json_file_map_user)

            with open(json_path_map_user, 'r') as jsonFile_map_user:
                data_map_user = json.load(jsonFile_map_user)

                try:
                    for map_user in data_map_user['data']['hoverData']:
                        district = map_user
                        registered_users = data_map_user['data']['hoverData'][map_user]['registeredUsers']
                        app_opens = data_map_user['data']['hoverData'][map_user]['appOpens']

                        map_user_result['State'].append(state)
                        map_user_result['Year'].append(year)
                        map_user_result['Quarter'].append('Q' + os.path.splitext(json_file_map_user)[0])
                        map_user_result['District'].append(district)
                        map_user_result['Registered_users'].append(registered_users)
                        map_user_result['App_Opens'].append(app_opens)
                except TypeError:
                    pass

df_map_user = pd.DataFrame(map_user_result)
```

```python
df_map_user.head()
```

```python
df_map_user.to_csv("PhonePe_Map_User.csv",index = False)
```

```python
top_trans_path = "/content/pulse/data/top/transaction/country/india/state"

top_trans_result = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Pincode': [],
    'Transaction_Count': [],
    'Transaction_Amount': []
}

state_top_trans = os.listdir(top_trans_path)

for state in state_top_trans:
    state_path_top_trans = os.path.join(top_trans_path, state)
    years_top_trans = os.listdir(state_path_top_trans)

    for year in years_top_trans:
        year_path_top_trans = os.path.join(state_path_top_trans, year)
        json_files_top_trans = os.listdir(year_path_top_trans)

        for json_file_top_trans in json_files_top_trans:
            json_path_top_trans = os.path.join(year_path_top_trans, json_file_top_trans)

            with open(json_path_top_trans, 'r') as jsonFile_top_trans:
                data_top_trans = json.load(jsonFile_top_trans)

                try:
                    for top_trans in data_top_trans['data']['pincodes']:
                        pincode = top_trans['entityName']
                        count = top_trans['metric']['count']
                        amount = top_trans['metric']['amount']

                        top_trans_result['State'].append(state)
                        top_trans_result['Year'].append(year)
                        top_trans_result['Quarter'].append('Q' + os.path.splitext(json_file_top_trans)[0])
                        top_trans_result['Pincode'].append(pincode)
                        top_trans_result['Transaction_Count'].append(count)
                        top_trans_result['Transaction_Amount'].append(amount)
                except TypeError:
                    pass

df_top_trans = pd.DataFrame(top_trans_result)
```
